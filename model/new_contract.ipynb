```python
#filling missing value
df['Assists']=df['Assists'].fillna(df['Assists'].median())
df['Injury_Days']=df['Injury_Days'].fillna(df['Injury_Days'].median())
df['Market_Value_M']=df['Market_Value_M'].fillna(df['Market_Value_M'].median())  #the market value refers to 18.2 miilion like that
```

```python
sns.boxplot(y='Minutes_Played',data=df)
```

```python
Q1=np.percentile(df['Minutes_Played'],25,method='midpoint')
Q3=np.percentile(df['Minutes_Played'],75,method='midpoint')
IQR=Q3 - Q1
print(IQR)
```

```python
lower=Q1-1.5*IQR
lower
```

```python
val=df['Minutes_Played'].median()
for i in df.index:
    if (df.loc[i,'Minutes_Played']<=lower):
        df.loc[i,'Minutes_Played']=val
```

```python
#Low playing time is a meaningful coaching decision, not an outlier error, so it was retained and explicitly modeled rather than imputed
```

```python
#feature_engineering
df['Attack_Contribution'] = df['Goals'] + df['Assists']

df['Trust_Adjusted_Usage'] = df['Minutes_Played'] * df['Coach_Trust_Score']

df['Injury_Risk_Index'] = df['Injury_Days'] / df['Age']

df['GK_Effectiveness'] = df['Saves'] / df['Minutes_Played']
```

```python
from sklearn.preprocessing import LabelEncoder

lab=LabelEncoder()
df['Position']=lab.fit_transform(df['Position'])
df['Continent']=lab.fit_transform(df['Continent'])
```

```python
x=df.drop(columns=['Contract_Renewed','Player_Name']).values
y=df['Contract_Renewed'].values
```

```python
x
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(n_estimators=100)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.predict([[30,3,2,2011,13,8.0,59,15.0,18.2,2,74,0,18.0,174122,0.517241,0]]))
```

```python
print(model.predict([[37,2,3,120,0,0.0,25,120.0,1.0,3,20,0,0.0,174122,0.03,0]]))
```

```python
from sklearn.metrics import accuracy_score,classification_report

score=accuracy_score(y_test,y_pred)
score
```

```python
report=classification_report(y_pred,y_test)
print(report)
```
